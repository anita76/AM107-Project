```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("ggplot")
%matplotlib inline
```

```python
#Accuracy scores in order of 

#Circles
circles_acc = [0.991, 0.992, 0.989, 0.991, 0.989]
moons_acc = [0.973, 0.913, 0.967, 0.973, 0.889]
overlap_acc = [0.823, 0.824, 0.818, 0.826, 0.824]

```

```python
labels =('Deep Ensemble', 'Dropout Sampling', 'Random Sampling', 'HMC', 'ADVI')

fig, ax = plt.subplots(figsize=(15,5))
#ax.yaxis.set_major_formatter(formatter)
plt.bar(np.arange(5), circles_acc, color='b')
plt.xticks(np.arange(5),labels )
ax.set_xticklabels(labels,rotation=0, fontsize=14)
plt.ylim(0.95,1)
plt.title("Circles Dataset Test Set Accuracy (Posterior Predictive Mean)")
plt.savefig("circles_accuracy_plot.png")
plt.show()
```

```python

fig, ax = plt.subplots(figsize=(15,5))
#ax.yaxis.set_major_formatter(formatter)
plt.bar(np.arange(5), moons_acc, color='b')
plt.xticks(np.arange(5),labels )
ax.set_xticklabels(labels,rotation=0, fontsize=14)
plt.ylim(0.85,1)
plt.title("Moons Dataset Test Set Accuracy (Posterior Predictive Mean)")
plt.savefig("moons_accuracy_plot.png")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,5))
#ax.yaxis.set_major_formatter(formatter)
plt.bar(np.arange(5), overlap_acc, color='b')
plt.xticks(np.arange(5),labels )
ax.set_xticklabels(labels,rotation=0, fontsize=14)
plt.ylim(0.75,0.85)
plt.title("Overlap Dataset Test Set Accuracy (Posterior Predictive Mean)")
plt.savefig("overlap_accuracy_plot.png")
plt.show()
```
